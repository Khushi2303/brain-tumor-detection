```python
dataset.isnull().sum()
```

```python
dataset["Age"].fillna(dataset["Age"].mean(),inplace=True)
```

```python
md=np.median(dataset["Age"])
```

```python
dataset["Age"].median()
```

```python
sns.histplot(x="Age", data=dataset, bins=[i for i in range (0,81,10)])
plt.plot([md for i in range(0,300)], [i for i in range (0,300)],c="red")
plt.show()
```

```python
mo=dataset["Age"].mode()[0]
```

```python
dataset["Age"].value_counts()
```

```python
sns.histplot(x="Age", data= dataset, bins=[i for i in range(0,81,10)])
plt.plot([mo for i in range(0,300)], [i for i in range(0,300)],c="purple")
plt.show()
```

```python
sns.histplot(x="Age", data = dataset, bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range (0,300)], [i for i in range (0,300)], c="green", label="mean")
plt.plot([md for i in range(0,300)], [i for i in range(0,300)], c="purple", label="median")
plt.plot([mo for i in range(0,300)], [i for i in range(0,300)], c="yellow", label="mode")
plt.legend()
plt.show()
```
