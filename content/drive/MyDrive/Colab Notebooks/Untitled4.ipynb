```python
import zipfile
import os

# Define correct paths based on os.listdir() output
zip_paths = ["tumor_dataset.zip.zip", "tumor_dataset1.zip.zip"]
 # Updated names
extract_path = "/content/dataset"

# Extract datasets
for zip_path in zip_paths:
    if os.path.exists(zip_path):  # Ensure file exists
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(extract_path)
        print(f"Extracted: {zip_path}")
    else:
        print(f"File not found: {zip_path}")
```
