```python
def extract_text_from_pdf(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text
```

```python
def ask_gemini_api(api_key, prompt):
    client = genai.Client(api_key=api_key)
    model = "gemini-2.0-flash-lite"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(text=prompt),
            ],
        ),
    ]

    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=40,
        max_output_tokens=512,  # Adjust based on needs
        response_mime_type="text/plain",
    )

    response = ""
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        response += chunk.text

    return response
```

```python
def answer_question_from_pdf(pdf_path, api_key, question):
    extracted_text = extract_text_from_pdf(pdf_path)
    prompt = f"Answer the following question based on the document text:\n{extracted_text}\n\nQuestion: {question}"
    answer = ask_gemini_api(api_key, prompt)
    return answer
```
